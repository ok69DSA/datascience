```python
import pandas as pd 
import numpy as np

df = pd.read_csv("C:/Users/Lenovo/Downloads/Toyota.csv")
print("Original:", df.head, df.shape)

numeric = df.select_dtypes(include=['int64', 'float64'])
print("Numerica data:", numeric.head, numeric.shape)

numeric_clean = numeric.dropna()
print("After Droping NA :",numeric_clean.head , numeric_clean.shape)

numeric_clean = numeric_clean.drop_duplicates()
print("After Droping Duplicates :",numeric_clean.head , numeric_clean.shape)


z_score = np.abs((numeric_clean - numeric_clean.mean()) / numeric_clean.std())
threshold = 3
df_clean = df.loc[numeric_clean[(z_score < threshold).all(axis=1)].index]  # Learn
print("After removing outliner ", df_clean.head()) 

print("\nCleaned Shape:", df_clean.shape)
```
