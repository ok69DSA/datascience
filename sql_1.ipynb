```python
import sqlite3

conn = sqlite3.connect("sample9.db")

curr = conn.cursor()

com1 = '''CREATE TABLE IF NOT EXISTS 
            student(prn INTEGER, name TEXT, grade TEXT)'''

curr.execute(com1)

com2 = "INSERT INTO student VALUES('1','Ram','A')"
curr.execute(com2)

conn.commit()

query = "SELECT * FROM student"
curr.execute(query)
row = curr.fetchall()

for rows in row:
    print(rows)

```
