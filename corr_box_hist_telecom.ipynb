```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np

df = pd.read_csv("churn-bigml-80.csv")
df.head()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only = True),fmt=".2f",annot = True,cmap="coolwarm")
plt.show()
```

```python
#Implementing on 1 col as not mentioned
sns.boxplot(data = df,x='Total day minutes')
plt.show()
```

```python
sns.histplot(data=df,x="Total day minutes",kde = True)
```
