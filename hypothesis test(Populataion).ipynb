```python
import pandas as pd 
import numpy as np
from scipy import stats
population=[69,85,96,69,75,88,20,25,65,66,67,28,30,50,45,34,65,15,12,10,11,18,19,17]
mean=np.mean(population)
print("mean of population :",mean)
sample=np.random.choice(population,24)

t_stat,p_value=stats.ttest_1samp(sample,mean)
if p_value < 0.05 :
    print("We reject NULL hyposthesis")
else :
    print("We Accept NULL hypostheses")    
```
