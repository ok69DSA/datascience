```python
#Creating a Text file
with open("test1.txt",'w') as f:
    #Contents can be chnaged
    f.write('''Hello
    My name is Shyam
    My name is Ram''')
    f.close()


with open("test1.txt",'r') as f:
    content = f.read()
    f.close()

#we were not able to read the last so added this
content += " "
a = ""
flag = False
word = []
count = []
for i in content:
    if(i.isalpha()):
        a+=i
    else:
        #if a is to check is 'a' is a real word or a space
        if a:
            for j in word:
                if (j == a):
                    idx = word.index(a)
                    count[idx] += 1
                    flag = True
                    break
            if(not flag):
                word.append(a)
                count.append(1)
            a = ""
            flag = False

for w,c in zip(word,count):
    print(f"Word: {w}, Count = {c}")
```
