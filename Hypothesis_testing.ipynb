```python
import scipy.stats as stats
import numpy as np
import random

pop_mean = 50
sample = [0] * 100

sample_size = len(sample)
for i in range(100):
    sample[i] = random.randint(0,100)

sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
alpha = 0.05

know_pop_std = None

if(know_pop_std is not None):
    z_score = (sample_mean - pop_mean)/(know_pop_std / np.sqrt(sample_size))
    p_value = 2 * (1-stats.norm.cdf(abs(z_score)))
    test_stat = z_score
    test = 'z'
else:
    t_stat = (sample_mean - pop_mean)/(sample_std / np.sqrt(sample_size))
    p_value = 2 * (1-stats.t.cdf(abs(t_stat),df=sample_size - 1))
    test_stat = t_stat
    test = 't'

print(f"Sample mean: {sample_mean}")
print(f"Population mean: {pop_mean}")
print(f"{test} statistic : {test_stat}")
print(f"p-value: {p_value}")

if(p_value < alpha):
    print("Null Hypothesis rejected")
else:
    print("Hypothesis Sucess")
```
